# Garantie du passage à l’échelle

## CLOUD

```python
import pandas as pd
```

```python
df = pd.read_csv('results/Garantie-du-passage-a-l-echelle_CLOUD_32_CORES.csv')
df = df.drop(columns=['Unnamed: 0'])
df['sf_train_time'] = df['train_time']/9.361921
df['sf_dataset_size'] = df['dataset_size']/360643033
df['dataset_size'] = df['dataset_size']/100000000
df['sf_difference'] = df['sf_dataset_size'] - df['sf_train_time']
df = df.round(2)
df
```

```python
dftemp = df[['sf_train_time', 'sf_dataset_size', 'sf_difference', 'dataset_size']]
dftemp.columns=['SF Training time', 'SF Dataset', 'SF Difference', 'Dataset Size (GB)']
dftemp.plot.line(
    x='Dataset Size (GB)', 
    xlabel="Dataset Size (GB)",
    ylabel="Scale factor",
    rot=0, 
    title='Scaling factor of dataset increase compared to training time', 
    figsize=(15,10), 
    fontsize=12)
```

```python
dftemp = df[['dataset_size', 'time', 'sf_dataset_size', 'sf_train_time', 'sf_difference', ]]
dftemp.columns=['Dataset Size (GB)', 'Training time (s)', 'SF Dataset', 'SF Training time', 'SF Difference', ]
dftemp
```
