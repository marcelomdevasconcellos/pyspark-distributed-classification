```python
import pandas as pd
```

```python
df = pd.read_csv('results/CLOUD_NUMBER_OF_CORES_32.csv')
```

```python
df
```

```python
df['dataset_size'] = df['dataset_size']/100000000
```

```python
df['dataset_size_num'] = df['dataset_size_num']/100
```

```python
df['sf_train_time'] = df['train_time']/9.361921
```

```python
df = df.round(2)
```

```python
dftemp = df[['sf_train_time', 'dataset_size_num', 'dataset_size']]
dftemp.columns=['Training time', 'Dataset', 'dataset_size']
dftemp.plot.line(
    x='dataset_size', 
    xlabel="Dataset Size (GB)",
    ylabel="Scale factor",
    rot=0, 
    title='Scaling factor of dataset increase compared to training time', 
    figsize=(15,10), 
    fontsize=12)
```
